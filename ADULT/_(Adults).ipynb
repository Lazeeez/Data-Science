```python
data = pd.read_csv("/content/drive/My Drive/adult.csv") 
data
```

```python
data = data.replace('?', np.nan)
data
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data.isnull().sum()
```

```python
#label encoding

from sklearn.preprocessing import LabelEncoder
l1 = LabelEncoder()

l1.fit(data['workclass'])
data.workclass = l1.transform(data.workclass)

l1.fit(data['education'])
data.education = l1.transform(data.education)

l1.fit(data['maritalstatus'])
data.maritalstatus = l1.transform(data.maritalstatus)

#l1.fit(data['occupation'])
#data.occupation = l1.transform(data.occupation)

l1.fit(data['relationship'])
data.relationship = l1.transform(data.relationship)

l1.fit(data['race'])
data.race = l1.transform(data.race)

l1.fit(data['gender'])
data.gender = l1.transform(data.gender)
gender = dict(zip(l1.classes_, l1.transform(l1.classes_)))
print(gender)

l1.fit(data['nativecountry'])
data.nativecountry = l1.transform(data.nativecountry)
native = dict(zip(l1.transform(l1.classes_), l1.classes_ ))
print(native)

l1.fit(data['income'])
data.income = l1.transform(data.income)

data
```

```python
#one-hot encoding

oh = pd.get_dummies(data['occupation'])
data = data.drop('occupation',axis = 1)
data = data.join(oh)

data.head()
```
