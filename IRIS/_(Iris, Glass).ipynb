```python
data = pd.read_csv("/content/drive/My Drive/iris.csv") 
print (data.describe())
print ("---------------------------------------------------")

Iset = data[data["class"]=="Iris-setosa"]
Iversi = data[data["class"]=="Iris-versicolor"]
Iverg = data[data["class"]=="Iris-virginica"]

print ("Iris-setosa\n")
print(Iset.describe())
print ("---------------------------------------------------")

print ("Iris-versicolor\n")
print(Iversi.describe())
print ("---------------------------------------------------")

print ("Iris-virginica\n")
print(Iverg.describe())
print("----------------------------------------------------")

plt.scatter(data["sepallength"], data["sepalwidth"], color='g')
plt.xlabel('sepal legth')
plt.ylabel('sepal width')
plt.show()

plt.scatter(Iset["sepallength"], Iset["sepalwidth"], color = 'r')
plt.scatter(Iversi["sepallength"], Iversi["sepalwidth"], color = 'g')
plt.scatter(Iverg["sepallength"], Iverg["sepalwidth"], color = 'b')
plt.xlabel('sepal legth')
plt.ylabel('sepal width')
plt.show()

plt.title('all')
plt.boxplot([data["sepallength"], data["sepalwidth"], data["petallength"], data["petalwidth"]])
plt.xticks((1,2,3,4), ('Sepal length', 'Sepal Width', 'Petal Length', 'Petal Width'))
plt.show()

plt.title('Iris setosa')
plt.boxplot([Iset["sepallength"], Iset["sepalwidth"], Iset["petallength"], Iset["petalwidth"]])
plt.xticks((1,2,3,4), ('Sepal length', 'Sepal Width', 'Petal Length', 'Petal Width'))
plt.show()

plt.title('iris versicolor')
plt.boxplot([Iversi["sepallength"], Iversi["sepalwidth"], Iversi["petallength"], Iversi["petalwidth"]])
plt.xticks((1,2,3,4), ('Sepal length', 'Sepal Width', 'Petal Length', 'Petal Width'))
plt.show()

plt.title('iris verginca')
plt.boxplot([Iverg["sepallength"], Iverg["sepalwidth"], Iverg["petallength"], Iverg["petalwidth"]])
plt.xticks((1,2,3,4), ('Sepal length', 'Sepal Width', 'Petal Length', 'Petal Width'))
plt.show()
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
df_out = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
(df_out)
```

```python
plt.boxplot([df_out["RI"]])
plt.grid(True)
plt.xticks((0,1,2), ('', 'RI', ''))
plt.show()

plt.boxplot([df_out["Na"]])
plt.grid(True)
plt.xticks((0,1,2), ('', 'Na', ''))
plt.show()

plt.boxplot([df_out["Mg"]])
plt.grid(True)
plt.xticks((0,1,2), ('', 'Mg', ''))
plt.show()

plt.boxplot([df_out["Al"]])
plt.grid(True)
plt.xticks((0,1,2), ('', 'Al', ''))
plt.show()

plt.boxplot([df_out["Si"]])
plt.grid(True)
plt.xticks((0,1,2), ('', 'Si', ''))
plt.show()

plt.boxplot([df_out["K"]])
plt.grid(True)
plt.xticks((0,1,2), ('', 'K', ''))
plt.show()

plt.boxplot([df_out["Ca"]])
plt.grid(True)
plt.xticks((0,1,2), ('', 'Ca', ''))
plt.show()

plt.boxplot([df_out["Ba"]])
plt.grid(True)
plt.xticks((0,1,2), ('', 'Ba', ''))
plt.show()

plt.boxplot([df_out["Fe"]])
plt.grid(True)
plt.xticks((0,1,2), ('', 'Fe', ''))
plt.show()
```
